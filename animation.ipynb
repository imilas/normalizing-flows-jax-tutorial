```python
# modified https://blog.evjang.com/2019/07/nf-jax.html
%matplotlib widget
from matplotlib import animation, rc
from IPython.display import HTML, Image
import matplotlib.pyplot as plt 
import numpy as np 

fig, ax = plt.subplots()
plt.ylim([-10, 10])
plt.xlim([-10, 10])

y = np.load("transformations.npy")
values = y
paths = ax.scatter(y[:, 0], y[:, 1], s=10, color='red')

def animate(i):
    l = i//48
    t = (float(i%48))/48
    y = (1-t)*values[l] + t*values[l+1]
    paths.set_offsets(y)
    return (paths,)
anim = animation.FuncAnimation(fig, animate, frames=48*len(values), interval=20,)
# anim.save('anim.gif', writer='pillow')
plt.show()
```
